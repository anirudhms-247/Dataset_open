```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
skills=pd.read_csv('./Employee_Skills_Datset.csv')

empdetails=pd.read_csv('./Employee_Designation.csv')
```

```python
names=empdetails["Ename"]
skills.drop('Eid',axis='columns', inplace=True)
entity_pairs=[]
G = nx.Graph()
for skill in skills:
    for (index,weight) in enumerate(skills[skill]):
        if(weight):
            G.add_edge(skill,names[index],weight = weight)
```

```python
plt.figure(figsize=(10,10))
pos=nx.spring_layout(G) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edges(G,pos,edge_color='c')
plt.show()
```
